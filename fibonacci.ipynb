```python
def recursiveFibonacci(n):
    if n==0 or n==1:
        return n
    else:
       return recursiveFibonacci(n-1)+recursiveFibonacci(n-2)
  
```

```python
recursiveFibonacci(5)
```

```python
def iterativeFibonacci(n):
    x ,y=0, 1
    for i in range(n):
       x,y=y, x+y
        
    return x
```

```python
 iterativeFibonacci(8)
```
